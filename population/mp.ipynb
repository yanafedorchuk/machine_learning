```python
#подгружаем необходимые библиотеки
import pandas as pd
import numpy as np
```

```python
cities=pd.read_csv('PLS.csv')
cities_new=cities.drop(['name'], axis=1)
cities_new
```

```python
a=cities['name89']
names=list(a)
names
```

```python
cities_new['D'] = 0
cit_n=cities_new.fillna(0)
cit_n
```

```python
d=[]
k=np.array(cit_n.pop10)
i=cit_n.Latitude[0]
for j in cit_n.Latitude:
    d=np.append(d,(i-j)**2)
b=[]
g=cit_n.Longitude[0]
for p in cit_n.Longitude:
    b=np.append(b,(g-p)**2)
sqrt=(b+d)**(1/2)
```

```python
cit_n['D']=sqrt
cit_n['C']=k
cols_to_convert = ['C','D']
for col in cols_to_convert:
    cit_n[col] = pd.to_numeric(cit_n[col], errors='coerce')
    cit_n['dist']=cit_n['D']/cit_n['C']
cit_n['dist'][cit_n['dist']>= 1E308] = 0
cit_n['dist'].sum()
```

```python
import mpu

def dist_calculator(lat1,lon1,lat2,lon2):
    dist = mpu.haversine_distance((lat1, lon1), (lat2, lon2))
    return(dist)
```

```python
dist_calculator(52.2296756,21.0122287,52.406374, 16.9251681)
```

```python
def mp_calculator1(lat1,lon1,lat2,lon2,pop):
    first_step=pop/dist_calculator(lat1,lon1,lat2,lon2)
    return first_step
```
