```python
#подгружаем необходимые библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import GridSearchCV
```

```python
# датасет с сайта kaggle
df=pd.read_csv('diamonds.csv')
df=df.drop(['Unnamed: 0'], axis=1)
cat_feats = ['cut', 'color', 'clarity']
lab_enc = LabelEncoder()
for i in range(3):
    new = lab_enc.fit_transform(df[cat_feats[i]])
    df[cat_feats[i]] = new
    
X = df[['carat', 'depth', 'table', 'x', 'y', 'z', 'clarity', 'cut', 'color']]
y = df[['price']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
regr = RandomForestRegressor(n_estimators = 10, max_depth = 10, random_state = 42)
regr.fit(X_train, y_train.values.ravel())
predict = regr.predict(X_test)
df_pred = X_test
df_pred['price'] = y_test
df_pred['prediction'] = predict.tolist()
df_pred.head(15)
```

```python
x_axis = X_test.carat
plt.scatter(x_axis, y_test, color = 'navy', alpha = 0.5, marker = '.', label = 'Real Data')
plt.scatter(x_axis, predict, color = 'orangered', alpha = 0.35, marker = '.', label = 'Prediction')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.grid()
plt.legend(loc = 'upper left')
plt.show()
```

#### Mean absolute error (MAE), Mean squared error (MSE), $R^{2}$

```python
mae = mean_absolute_error(y_test.values.ravel(), predict)
mse = mean_squared_error(y_test.values.ravel(), predict)
r2 = r2_score(y_test.values.ravel(), predict)
print('Mean Absolute Error:', round(mae, 2))
print('Mean Squared Error:', round(mse, 2))
print('R-squared:', round(r2, 2))
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {
    'max_depth': [60,70, 80],
    'n_estimators': [900, 1000, 1100]
}
gridforest = GridSearchCV(regr, parameters, cv = 3, n_jobs = -1, verbose = 1)
gridforest.fit(X_train, y_train)
gridforest.best_params_
```
